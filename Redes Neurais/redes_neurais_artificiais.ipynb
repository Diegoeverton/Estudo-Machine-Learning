```python
from sklearn.neural_network import MLPClassifier
```

```python
import pickle
```

```python
with open(r'D:\CURSOS\UDEMY\Machine Learning com Python\codigos\credit.pkl', 'rb')  as f:
    X_credit_treinamento, y_credit_treinamento, X_credit_teste, y_credit_teste = pickle.load(f)
```

```python
print(X_credit_treinamento.shape)
print(y_credit_treinamento.shape)
```

```python
print(X_credit_teste.shape)
print(y_credit_teste.shape)
```

```python
rede_neural_credit = MLPClassifier(max_iter=1500, verbose=True, tol=0.0000100, 
                                   solver='adam', activation='relu', 
                                   hidden_layer_sizes=(50, 50), random_state=0)
rede_neural_credit.fit(X_credit_treinamento, y_credit_treinamento)
```

```python
previsoes = rede_neural_credit.predict(X_credit_teste)
```

```python
previsoes
```

```python
y_credit_teste
```

```python
from sklearn.metrics import accuracy_score, classification_report

accuracy_score(y_credit_teste, previsoes)
```

```python
print(classification_report(y_credit_teste, previsoes))
```

```python
from yellowbrick.classifier import ConfusionMatrix
cm = ConfusionMatrix(rede_neural_credit)
cm.fit(X_credit_treinamento, y_credit_treinamento)
cm.score(X_credit_teste,y_credit_teste)
```

```python
X_credit_treinamento, y_credit_treinamento, X_credit_teste, y_credit_teste
```

```python
import numpy as np
X_credit = np.concatenate(((X_credit_treinamento, X_credit_teste)), axis=0)
y_credit = np.concatenate(((y_credit_treinamento, y_credit_teste)), axis=0)
```

```python

from sklearn.model_selection import GridSearchCV

parametros = {'activation':['relu', 'logistic', 'tahn'],
              'solver': ['adam', 'sigmoid'],
              'batch_size': [10, 56]}

grid_search = GridSearchCV(estimator=MLPClassifier(), param_grid=parametros)
grid_search.fit(X_credit, y_credit)
melhopres_parametros = grid_search.best_params_
melhopres_resultados = grid_search.best_score_
print(melhopres_parametros)
print(melhopres_resultados)
```
