```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.svm import SVC
```

### base dados credit

```python
import pickle
with open('credit.pkl', 'rb') as f:
    X_credit_treinamento, y_credit_treinamento, X_credit_teste, y_credit_teste = pickle.load(f)
```

```python
print(X_credit_treinamento.shape)
print(y_credit_treinamento.shape)
```

```python
print(X_credit_teste.shape)
print(y_credit_teste.shape)
```

```python
svm_credit = SVC(kernel='rbf', random_state=1, C=2.0)
svm_credit.fit(X_credit_treinamento, y_credit_treinamento)
```

```python
previsoes = svm_credit.predict(X_credit_teste)
previsoes
```

```python
y_credit_teste
```

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy_score(y_credit_teste, previsoes)
```

```python

print(classification_report(y_credit_teste, previsoes))
```

```python
from yellowbrick.classifier import  ConfusionMatrix
cm = ConfusionMatrix(svm_credit)
cm.fit(X_credit_treinamento, y_credit_treinamento)
cm.score(X_credit_teste, y_credit_teste)
```

Ajuste de parametros com gridsearch 

```python
import numpy as np
```

```python
X_credit = np.concatenate(((X_credit_treinamento, X_credit_teste)), axis=0)
y_credit = np.concatenate(((y_credit_treinamento, y_credit_teste)), axis=0)
```

```python
from sklearn.model_selection import GridSearchCV

parametros = {'tol': [0.001, 0.0001, 0.00001],
              'C': [1.0, 1.5, 2.0],
              'kernel': ['rbf', 'linear', 'poly', 'sigmoid']}

grid_search = GridSearchCV(estimator=SVC(), param_grid=parametros)
grid_search.fit(X_credit, y_credit)
melhopres_parametros = grid_search.best_params_
melhopres_resultados = grid_search.best_score_
print(melhopres_parametros)
print(melhopres_resultados)
```
